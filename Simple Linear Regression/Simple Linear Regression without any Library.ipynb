#    Simple Linear Regression of randomly generated points without using any Regression Library

Step 1: Import some necessary Libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

Step 2: Randomly generate datapoints for x and y

```python
# randomly genertaed input feature(x) between 1 to 100 of size:100.
x = [random.randint(1,100) for _ in range(100)]

# randomly genertaed output feature(y) between 1 to 100 of size:100.
y = [random.randint(1,100) for _ in range(100)]
print(x)
print(y)
```

Step 3: Calculate the means

```python
x_mean = sum(x)/len(x)
y_mean = sum(y)/len(y)
```

Step 4: Calculate the deviations

```python
x_deviations = [x - x_mean for x in x]
y_deviations = [y - y_mean for y in y]
```

Step 5: Calculate the sum of products

```python
sum_xy = sum(x * y for x, y in zip(x_deviations, y_deviations))
```

Step 6: Calculate the sum of squared deviations

```python
sum_x_squared = sum(x ** 2 for x in x_deviations)
```

Step 7: Calculate the regression coefficients

```python
slope = sum_xy / sum_x_squared
intercept = y_mean - slope * x_mean
```

Step 8: Calculation of prediction

```python
def predict(x):
    return intercept + slope * x
```

Step 9: Calculate errors

```python
errors = [y - predict(x) for x, y in zip(x,y)]
mean_squared_error = sum(error ** 2 for error in errors) / len(errors)
root_mean_squared_error = mean_squared_error ** 0.5
```

Step 10: Plot the data points and regression line

```python
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, [predict(x) for x in x], color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

Residual Plot: A residual plot shows the difference between the observed y-values and the predicted y-values. It helps to assess the goodness of fit and identify any patterns or trends in the residuals.

```python
residuals = [y - predict(x) for x, y in zip(x, y)]
plt.scatter(x, residuals, color='green')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

Actual vs. Predicted Plot

```python
plt.scatter(y, [predict(x) for x in x], color='blue')
plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linestyle='--')
plt.xlabel('Actual y')
plt.ylabel('Predicted y')
plt.title('Actual vs. Predicted')
plt.show()
```

Step 11: Display the errors

```python
print("Errors:")
for i, error in enumerate(errors):
    print(f"Data Point {i+1}: {error}")
print('-----------------------------------------------------------------')
print(f"Mean Squared Error: {mean_squared_error}")
print(f"Root Mean Squared Error: {root_mean_squared_error}")
```

Summary:
In this analysis, we applied simple linear regression to randomly generated data points. However, we observed that simple linear regression is not suitable for such randomly generated data. The nature of the data is not fixed, and the errors in the predictions are high and unpredictable.

Upon examining the data, we found that the relationship between the independent variable (x) and the dependent variable (y) was not well captured by a straight line. The errors, represented by the residuals, exhibited no clear pattern or trend. This indicates that a linear regression model is not appropriate for this type of data.

The scatter plot with the regression line showed a weak or no linear relationship between x and y. The actual vs. predicted plot also revealed significant deviations from the ideal scenario, with the predicted values often deviating from the actual values.

It is important to note that simple linear regression assumes a linear relationship between the variables and assumes that the errors are normally distributed with constant variance. These assumptions may not hold for randomly generated data, resulting in poor model performance and high prediction errors.

In conclusion, for randomly generated data points without a fixed underlying relationship, it is advisable to explore other regression techniques or alternative approaches to capture the complex nature of the data accurately.
